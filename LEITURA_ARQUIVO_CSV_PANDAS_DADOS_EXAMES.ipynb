# AJUSTAR O SCHEMA DOS DADOS, SE NECESSÁRIO

```python
df['Cidade'] = df['Cidade'].astype('string')
df['Estado'] = df['Estado'].astype('string')
df['Bairro'] = df['Bairro'].astype('string')
df['NumPedidoMedico'] = df['NumPedidoMedico'].astype('string')
df['Exame'] = df['Exame'].astype('string')
df['SiglaExame'] = df['SiglaExame'].astype('string')
df['Material'] = df['Material'].astype('string')
df['SetorExame'] = df['SetorExame'].astype('string')
df['PrecoExame'] = df['PrecoExame'].str.replace(',','.').astype('double')
df['DataPrevistaResultado'] = df['DataPrevistaResultado'].astype('datetime64[ns]')
df['DataLiberacaoResultado'] = df['DataLiberacaoResultado'].astype('datetime64[ns]')

df.info()
```

```python
# Tabela de Unidades
unidades = df[['IdUnidadeAtendimento', 'Cidade', 'Estado', 'Bairro']].drop_duplicates()

# Tabela de Exames
exames = df[['IdExame', 'SiglaExame', 'Material']].drop_duplicates()

# Tabela de Pedidos
pedidos = df[['NumPedidoMedico', 'DataPrevistaResultado', 'DataLiberacaoResultado']].drop_duplicates()

# Tabela de Setores
setores = df[['SetorExame', 'PrecoExame', 'QtdExames']].drop_duplicates()

# Exibir tabelas normalizadas
print("\nTabela de Unidades:")
print(unidades.head())

print("\nTabela de Exames:")
print(exames.head())

print("\nTabela de Pedidos:")
print(pedidos.head())

print("\nTabela de Setores:")
print(setores.head())

df.head(5)
```
